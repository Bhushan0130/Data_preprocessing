# Data Cleaning

### Missing Value imputation by Mean, Median

```python
# load Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
path = 'https://raw.githubusercontent.com/Bhushan0130/Pratice_project/main/houseprice_train.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
pd.set_option( 'display.max_columns', None)
pd.set_option ('display.max_rows', None)
df.head(2)
```

```python
nullvalue_per = df.isnull().sum() / df.shape[0] * 100
nullvalue_per
```

```python
missing_value_clm_gre_20 = nullvalue_per[nullvalue_per > 20]
```

```python
missing_value_clm_gre_20
```

```python
missing_value_clm_gre_20 = nullvalue_per[nullvalue_per > 20].keys()
```

```python
missing_value_clm_gre_20
```

```python
df2 = df.drop(columns = missing_value_clm_gre_20)
df2.shape
```

```python
numeric_df = df2.select_dtypes(include = ['int', 'float'])
numeric_df.head()
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(numeric_df.isnull())
```

```python
numeric_df[numeric_df.isnull().any(axis=1)]
```

```python
numeric_df.isnull().sum()
```

```python
missing_var_name = [var for var in numeric_df.columns if numeric_df[var].isnull().sum() > 0]
missing_var_name
```

```python
plt.figure(figsize = (25, 25))
sns.set()

for i, var in enumerate(missing_var_name):
    plt.subplot(2, 2, i+1)    
    sns.distplot(numeric_df[var], bins = 20, kde_kws = { 'linewidth' : 5, 'color' : '#DC143C'})
```

```python
numeric_df2 = numeric_df.fillna(numeric_df.mean())
numeric_df2.isnull().sum().sum()
```

```python
plt.figure(figsize = (25, 25))
sns.set()

for i, var in enumerate(missing_var_name):
    plt.subplot(2, 2, i+1)
    sns.distplot(numeric_df[var], bins = 20, kde_kws = {'linewidth' : 5, 'color' : 'red' }, label = "Original")
    sns.distplot(numeric_df2[var], bins = 20, kde_kws = {'linewidth' : 5, 'color' : 'green' }, label = 'Mean')
    plt.legend()
```

```python
numeric_df3 = numeric_df.fillna(numeric_df.median())
numeric_df3.isnull().sum().sum()
```

```python
plt.figure(figsize = (25, 25))
sns.set()

for i, var in enumerate(missing_var_name):
    plt.subplot(2, 2, i+1)
    sns.distplot(numeric_df[var], bins = 20, hist = False, kde_kws = {'linewidth' : 5,  'color' : 'red' }, label = 'Original')
    sns.distplot(numeric_df2[var], bins = 20, hist = False, kde_kws = {'linewidth' : 5, 'color' : 'green' }, label = 'Mean')
    sns.distplot(numeric_df3[var], bins = 20, hist = False, kde_kws = {'linewidth' : 5,  'color' : 'k' }, label = 'Median')
    plt.legend()
```

```python
# Note : if we find differene in mean and median then it mean many outlier is present in our dataframe or data
```
